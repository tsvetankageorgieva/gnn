```python
import pandas as pd
df = pd.read_csv("patients.csv")
print(df.head())
```

```python
X = df.drop(columns=["ID", "Outcome"])
y = df["Outcome"]
```

```python
from sklearn.neighbors import kneighbors_graph
# матрица на съседство 
# Използва се k-nn
A = kneighbors_graph(
    X,
    n_neighbors=2,
    mode="connectivity",
    include_self=False )
print(A)
```

```python
from sklearn.neighbors import radius_neighbors_graph
A = radius_neighbors_graph(
     X, 
     radius=5.0,  
     metric="euclidean",
     mode="connectivity")
print(A)
```

```python
import torch
row, col = A.nonzero()
# преобразуванане в списък от ребра
edge_index = torch.tensor([row, col], dtype=torch.long )
print(edge_index)
```

```python
# Преобразуване на характеристиките в тензор
x = torch.tensor(X.values, dtype=torch.float)
x.shape
```

```python
# Преобразуване на класовете
y = torch.tensor(y.values, dtype=torch.long)
print(y)
```

```python
# Създаване на графа
from torch_geometric.data import Data
data = Data( x=x, edge_index=edge_index, y=y)
print(data)
```

```python
# Записване във файл чрез съхраняване на отделните компоненти
torch.save({
    "x": data.x,
    "edge_index": data.edge_index,
    "y": data.y
}, "patients_graph.pt")
```

```python
# Зареждане от файл
import torch
graph = torch.load("patients_graph.pt")

from torch_geometric.data import Data

data = Data(
    x=graph["x"],
    edge_index=graph["edge_index"],
    y=graph["y"]
)
print(data)
```
